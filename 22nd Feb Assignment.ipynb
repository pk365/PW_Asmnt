#### Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.


**Ans.** In Pandas, a DataFrame is a two-dimensional labeled data structure with columns of potentially different data types. It is like a spreadsheet or a SQL table, where each column can have a different data type such as integers, floats, and strings. On the other hand, a Series is a one-dimensional labeled array that can hold any data type.

Here's an example to illustrate the difference between a Pandas DataFrame and a Series:

```python
import pandas as pd

# Example 1: Creating a DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Claire'],
    'Age': [25, 30, 27],
    'Gender': ['Female', 'Male', 'Female']
}
df = pd.DataFrame(data)
print(df)

# Example 2: Creating a Series
names = pd.Series(['Alice', 'Bob', 'Claire'])
print(names)
```

#### Q5. What are some common functions you can use to manipulate data in a Pandas DataFrame? Can you give an example of when you might use one of these functions?

**Ans.** There are many functions available in Pandas to manipulate data in a DataFrame. Here are a few commonly used functions:

**head()** and **tail():** Used to display the top or bottom n rows of the DataFrame.
Example:


```python
df = pd.read_excel('Data-for-Cleaning.xlsx')
df.head(10) # Display the top 10 rows of the DataFrame
```

**describe()**: Used to generate a statistical summary of the DataFrame.

```python
df = pd.read_excel('Data-for-Cleaning.xlsx')
df.describe()
```

**drop():** Used to drop rows or columns from the DataFrame.
Example:

```python
df = pd.read_excel('Data-for-Cleaning.xlsx')
df=df.drop("Last name",axis=1)
df
```

#### Q7. Create a DataFrame using multiple Series. Explain with an example.

```python
import pandas as pd

# create the first series
names = pd.Series(['Alice', 'Bob', 'Claire'])

# create the second series
ages = pd.Series([25, 30, 27])

# create the third series
genders = pd.Series(['Female', 'Male', 'Female'])

# create the DataFrame using the above series
df = pd.DataFrame({'Name': names, 'Age': ages, 'Gender': genders})

# print the DataFrame
print(df)
```
